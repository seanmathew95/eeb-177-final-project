# Column extractor

**file must be a formatted csv file with the head removed and any "\r" characters replaced with newlines**

for this particular file, "panthera.csv", I used the following Unix command

tail -n +1 > formatted-panthera.csv


This function will take any column and group them with the species name in a dictionary


```python

def column_extractor(file_name, column):
    file_input = open(file_name) #opens the inputted file
    file_list = file_input.readlines() #creates a list with each line in the file as a separate object
    extract_dict = {} #empty dictionary
    for line in file_list:
        species = line.split(",")[5] #species name is the sixth item in the list
        extract =[] #empty list to hold all values in the column of the same species
        for ll in file_list: #go through each line again and populate extract 
            if (ll.split(",")[5] == species): #only add items with the species currently being observed
                item = (ll.split(",")[column])
                extract.append(item)
        extract_dict[species] = extract
    for line in file_list:
        extract
    return extract_dict
```

```python
column_extractor("formatted-panthera.csv", 4) #extract the fourth column
```
